## Import The Libarries

```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.autograd import Variable
```

## Define the hyper parmiters

```python
input_size = 784   # number of input neurons (image pixels)
hidden_size = 400  # Number of hidden neurons
out_size = 10      # Number of classes (0-9)
epochs = 10        #How many times we pass our entire dataset into our network
batch_size = 100   #Input size of the data during the iteration
learning_rate = 0.001 # How fast we are
```

## loading our data

```python
# loading our data
train_dataset = datasets.MNIST(root='./data', train = True,
                              transform=transforms.ToTensor(),
                              download=True)

test_dataset = datasets.MNIST(root='./data', train = False,
                             transform=transforms.ToTensor())
```

# Create The Dataloader for training and Testing

```python
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                          batch_size=batch_size,
                                              shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                         batch_size=batch_size,
                                         shuffle=False)
```

## Build the Neural Network model

```python
#building the network
class Net(nn.Module):
    def __init__(self,input_size ,hidden_size ,out_size ):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(input_size,hidden_size) # frist layer
        self.relu = nn.ReLU() # activation function 
        self.fc2 = nn.Linear(hidden_size,hidden_size) # second layer
        self.fc3 = nn.Linear(hidden_size,out_size) # last layer (output layer)
        
    #Forward propagtion
    def forward(self ,x):
        out = self.fc1(x)
        out = self.relu(out)       
        out = self.fc2(out)
        out = self.relu(out)       
        out = self.fc3(out)
        return out        
```

```python
# create an object from the class 
net = Net(input_size , hidden_size , out_size)
CUDA = torch.cuda.is_available()
if CUDA :
    net = net.cuda()
# the loss fumnction --> Cross entropy loss
# It comes along with softmax

criterion = nn.CrossEntropyLoss()
optmizer = torch.optim.Adam(net.parameters(),lr=learning_rate)
```

## Training The Model

```python
correct_train = 0
total_train = 0
for epoch in range(epochs):
    for i , (images,labels) in enumerate (train_loader):
        #flatten images and reshape it
        images = Variable(images.view(-1,28*28))
        labels = Variable(labels) # labels --> actual output
        
        #clear the param_grad
        optmizer.zero_grad()
        outputs = net.forward(images)             # apply forward propgation
        _,predicted = torch.max(outputs.data,1) # store the second argument 
        
        total_train += labels.size(0)     # encriment total train size
        correct_train += (predicted==labels).sum() # predicted vs actual label
        
        # calculate the loss between predicted output and actual output
        loss = criterion(outputs, labels)
        loss.backward()                   # backpropgation
        optmizer.step()                   # update the wieghts
        
        if (i+1) %100 == 0:
            print('Epoch :{}/{} , Iteration {}/{}  ,Training loss: {} , Training Acurrcy:{}%'
                  .format(epoch+1 , epochs , i+1 , len(train_dataset)//batch_size , loss.data ,
                          (100*correct_train/total_train)))
print("DONE TRAINING!")
```

## Testing the model

```python
correct = 0
total = 0 
for images,labels in test_loader:
    images = Variable(images.view(-1,28*28))
    labels = Variable(labels)
    
    # apply forward propgate
    ouputs= net(images)
    _,predicted = torch.max(ouputs.data ,1)
    
    # increment total 
    total += labels.size(0)
    
    # calculate thr correect prediction
    correct = (predicted==labels).sum()
    
    # Calculate the accurcy
print(" the final test accurcy is {}".format((100 * correct / total )))
```
